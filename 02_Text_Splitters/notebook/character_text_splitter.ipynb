```python
from langchain_text_splitters import CharacterTextSplitter
```

```python
text = """
Artificial intelligence is transforming technology and shaping the future. 
Machine learning algorithms are becoming more sophisticated every day.
Deep learning models can now process vast amounts of data efficiently.

Natural language processing has made significant strides in recent years.
Computer vision systems can now identify objects with remarkable accuracy.
Reinforcement learning is enabling robots to learn complex tasks autonomously.

The impact of AI extends across multiple industries including healthcare, finance, and transportation.
Ethical considerations around AI development are becoming increasingly important.
Researchers are working on making AI systems more transparent and explainable.
"""
```

```python
text_splitter = CharacterTextSplitter(chunk_size = 100, 
                                      chunk_overlap= 0, 
                                      length_function = len,
                                      separator = "")

chunks = text_splitter.split_text(text)
```

```python
chunks
```

```python
len(chunks)
```

On the bassis of space

```python
text_splitter = CharacterTextSplitter(chunk_size = 100, 
                                      chunk_overlap= 0, 
                                      length_function = len,
                                      separator = " ")

chunks = text_splitter.split_text(text)
chunks
```

```python
len(chunks)
```

On the bassis of paragraph

```python
text_splitter = CharacterTextSplitter(chunk_size = 100, 
                                      chunk_overlap= 0, 
                                      length_function = len,
                                      separator = "\n\n")

chunks = text_splitter.split_text(text)
chunks
```

```python
len(chunks)
```

chunk overlap

```python
text_splitter = CharacterTextSplitter(chunk_size = 100, 
                                      chunk_overlap= 23, 
                                      length_function = len,
                                      separator = "")

chunks = text_splitter.split_text(text)
chunks
```

### Token based splitting

```python
token_splitters = CharacterTextSplitter.from_tiktoken_encoder(
    encoding_name="cl100k_base",
    chunk_size = 50,
    chunk_overlap = 5
)

chunks = token_splitters.split_text(text)
```

```python
chunks
```

### **Examples Token vs Words**

Sentence:
`"I love artificial intelligence"`

Words: 4 words

Tokens may be:
`["I", " love", " artificial", " intelligence"]`

But not always. Important Point-- One word ≠ one token:

**Example 1-**

Word: `"unbelievable"`

Tokenizer may split it into: `["un", "believ", "able"]`

So: 1 word = 3 tokens

**Example 2-**

Sentence: `"What's up?"`

Tokens may become:`["What", "'s", " up", "?"]`

Splitting Documents

```python
from langchain_core.documents import Document

# create documents

docs = [Document(
    page_content=text,
    metadata={"source": "Text on AI"}
)]
```

```python
docs
```

```python
# create splitter

splitter = CharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=20,
    separator=""
)
```

```python

# split the documents

chunks = splitter.split_documents(docs)
```
