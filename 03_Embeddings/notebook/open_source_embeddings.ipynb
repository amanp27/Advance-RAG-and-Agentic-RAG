## Ollama

```python
query = "What is Openclaw/Moltbot and what are the major security concerns regarding this tool"
```

```python
query_embedding = ollama.embed(model = "embeddinggemma", input=query)
```

```python
query_embedding['embeddings']
```

```python
# dimensions

print(len(query_embedding["embeddings"][0]))
```

```python
query_embeddings = ollama.embed(
    model="embeddinggemma",
    input=query,
    dimensions=512
)
```

```python
# dimensions

print(len(query_embeddings["embeddings"][0]))
```

```python
# create the loader

loader = PyPDFLoader(file_path="../Openclaw_Research_Report.pdf")

docs = loader.load()

len(docs)
```

```python
# split documents

chunker = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=50
)

chunks = chunker.split_documents(docs)

len(chunks)
```

```python
# embed documents

text_documents = [doc.page_content for doc in chunks]

print(len(text_documents))
```

```python
# embed documents

document_embeddings = ollama.embed(
    model="embeddinggemma",
    input=text_documents
)
```

```python

document_embeddings = document_embeddings["embeddings"]
```

```python
# number of embedding vectors

print(len(document_embeddings))
```

```python

# output dimensions

print(len(document_embeddings[0]))
```

## Langchain Ollama

```python
langchain_embedder = OllamaEmbeddings(model = "embeddinggemma")
```

```python
langchain_document_embeddings = langchain_embedder.embed_documents(text_documents)
```
