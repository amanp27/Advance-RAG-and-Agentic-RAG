### Recurssive Loader

```python
from langchain_community.document_loaders import RecursiveUrlLoader
```

```python
base_url = "https://docs.langchain.com/oss/python/integrations/document_loaders"

loader = RecursiveUrlLoader(url = base_url)
```

```python
docs = loader.load()
```

```python
len(docs)
```

```python
for i in range(5):
    doc = docs[i]
    print(doc.metadata)
```

```python
documents_lazy_load = loader.lazy_load()
```

```python
counter = 0

for doc in documents_lazy_load:
    if counter == 10:
        break

    counter += 1

    pp(doc.page_content[:200])
    pp(doc.metadata)
```
