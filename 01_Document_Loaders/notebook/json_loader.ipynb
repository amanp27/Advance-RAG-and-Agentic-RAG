```python
from langchain_community.document_loaders.json_loader import JSONLoader
from pathlib import Path
```

```python
file_path = Path("../knowledge_source/apparels.json")
print(file_path.exists())
```

```python
loader = JSONLoader(file_path.as_posix(), jq_schema=".products[]", content_key= "Description")
docs = loader.load()
```

```python
docs
```

```python
for doc in docs:
    print(doc.page_content, end="\n\n")
```

```python
def metadata_func(record: dict, metadata: dict) -> dict:
    metadata["product_name"] = record["productName"]
    metadata["category"] = record["category"]

    return metadata
```

```python
loader = JSONLoader(file_path.as_posix(), jq_schema=".products[]", content_key= "Description", metadata_func= metadata_func)
docs = loader.load()
```
